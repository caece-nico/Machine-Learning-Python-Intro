```python
"""
Ene ste ejemplo creamos un grafico de barras para analizar la distribucion de las cantidades para predecir la 
bariable categorica.
Para esto usamos un groupby() y creamos una variable dummy que usamos para acumular.
Esta es una de las funciones mas complicadas.

1) usando np.ones creamos una nueva columna dummy con 1.
2) hacemos un groupby(count) por la columna dummy, la columna que vamos a usar para segmentar(categorica predictiva) y 
   la columna a visualizar. --> se guarda en un nuevo dataset totales.
3) Creamos una figura fig = plt.figure(figsize(10,7)) para despues crear un subplot    
4) De totales creamos DF temporal para bad_credit (variable a predecir) 1 y 2.
5) en el subplot plt.subplot(1,2,1) creamos un plt.bar(temp[col],temp.dummy)
6) hacemos lo mismo para bad_credit =2 en el subpot plt.subplot(1,2,2)

"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
import numpy as np
```

```python
url = 'German_credit.csv'
columnas = ['customer_id',
                  'checking_account_status', 'loan_duration_mo', 'credit_history', 
                  'purpose', 'loan_amount', 'savings_account_balance', 
                  'time_employed_yrs', 'payment_pcnt_income','gender_status', 
                  'other_signators', 'time_in_residence', 'property', 'age_yrs',
                  'other_credit_outstanding', 'home_ownership', 'number_loans', 
                  'job_category', 'dependents', 'telephone', 'foreign_worker', 
                  'bad_credit']
data = pd.read_csv(url, header= None, names = columnas)
data.head(5)
```

```python
def crea_bar_plot(data, columnas, x_col = 'bad_credit'):
    data['dummy']=np.ones(shape = data.shape[0]) ## creamos una columna con 1 en el DF
    for col in columnas:
        totales = data[['dummy','bad_credit',col]].groupby(['bad_credit',col], as_index = False).count() ## agrupamos por col y var a predecit
        temp = totales[totales['bad_credit']==1][[col,'dummy']] ## primero segmentamos por un vaor de la categoria
        fig = plt.figure(figsize = (10,7)) ## creamos la figura general, lo luego sera dividida en dos partes.
        plt.subplot(1,2,1) 
        plt.bar(temp[col],temp.dummy) ## creamos el barplot para la primera clase
        plt.title('bad_credit = 1')
        plt.xlabel(col)
        plt.ylabel('total')
        plt.subplot(1,2,2) ## usamos el segundo espacio
        temp = totales[totales['bad_credit']==2][[col,'dummy']] ## creamos el bar plot para la segunda clase.
        plt.bar(temp[col],temp.dummy)
        plt.title('bad_credit = 2')
        plt.xlabel(col)
        plt.ylabel('total')
        plt.show()
        plt.show()

        
        
```

```python
columns = ['checking_account_status', 'credit_history', 'purpose', 'savings_account_balance', 
                  'time_employed_yrs', 'gender_status', 'other_signators', 'property', 
                  'other_credit_outstanding', 'home_ownership', 'job_category', 'telephone', 
                  'foreign_worker']
crea_bar_plot(data,columns)
```
