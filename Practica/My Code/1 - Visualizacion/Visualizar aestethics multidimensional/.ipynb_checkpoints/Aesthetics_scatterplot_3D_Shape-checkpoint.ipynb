```python
"""
CREAMOS UNA FUNCION PARA PODER VISUALIZAR GRAFICOS EN 3D USANDO LA TECNICA AESTETHICS.
EN ESTE CASO ES: DOS VARIABLES + FORMA.
"""

"""
para hacer esto hay que tener en cuenta que va a crear un grafico por vez y luego los une, por eso em importante saber donde poner 
sea.set_style('whitegrid') y los plt.title, xlabel(), ylabel(), legen() y show().


IMPORTANTE: Para crear este grafico:
1) primero tenemos que tener en claro las dos variables numericas y la categorica.
2) dentro de la funcion definir una lista de formas.
3) busca las categorias unicas para la categoria_shape dentro del dataset.
4) loop por columnas.
5) loop enumerando idx y categoria (cat) del punto 3
6) crear un dataset tempral para una categoria por vez.
7) usar sea.regplot(col, y_col, dataset = temp, make = shape[idx], label = cat, scatter_kws{'alpha':}, fit_reg = False, color = '')
8) plt.legend()
"""
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sea
```

```python
url = 'Automobile price data _Raw_.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
## creamos la funcion limpia datos

def limpia_datos(data):
    import numpy as np
    import pandas as pd
    
    columnas = data.columns
    data.columns = [str.replace('-','_') for str in columnas]
    
    cols = ['price', 'bore', 'stroke', 
          'horsepower', 'peak_rpm']
    
    for c in cols:
        data.loc[data[c]=='?',c]=np.nan
        data[c].dropna(axis=0,inplace= True)
        
    for c in cols:
        data[c] = pd.to_numeric(data[c])
        
    return data
```

```python
limpia_datos(data)
```

```python
###creamos una funcion para graficar tres variables en un grafico.

def crea_3d_plot(dataset, x_col , y_col = 'price', shape_col ='fuel_type' , alpha = 0.2):
    shape = ['+','o','x','*'] ##definimos las formas, solo se usan tantas como categrias unicas existan.
    unique_cat = dataset[shape_col].unique() ## de nuestra columna categoria buscamos las categorias unicas.

    for col in x_col: ## hacemos un loop por columnas
        sea.set_style('whitegrid') ## por cada columna vamos a crear un grafico  
        for i, cat in enumerate(unique_cat): ## enumercamos las categorias 
            temporal = dataset[dataset[shape_col]==cat] ## para cada categoria egmentamos un DataSet
            sea.regplot(col, y_col, data = temporal, marker = shape[i], label = cat,
                        scatter_kws={'alpha':alpha},fit_reg = False, color = 'blue') ## crea el grafico para cada Cat de la columna,
                                                                                     ## si la columna tenia 3 categorias esto corre 3 veces
                                                                                     ## y generea un solo scatter con las 3 categorias.
        plt.title('Scatter plot de: '+ y_col + ' vs: ' + col)
        plt.xlabel(col)
        plt.ylabel(y_col)
        plt.legend()
        plt.show()
```

```python
num_cols = ['curb_weight', 'engine_size', 'horsepower', 'city_mpg']
crea_3d_plot(data, num_cols)                
```
