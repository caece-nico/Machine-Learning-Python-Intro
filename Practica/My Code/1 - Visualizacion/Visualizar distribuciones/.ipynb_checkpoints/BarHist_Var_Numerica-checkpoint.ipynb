```python
"""
En este ejemplo creamos una fucion para visualizar la 
distirbucion de alunas variales numericas pero teniendo en cuenta que si tienen algun string debe ser casteado o eliminado
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

url = 'Automobile price data _Raw_.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
# Creamos una copia de data
data_orig = data
```

```python
def elimina_guiones(data):
    columnas = data.columns
    data.columns = [str.replace('-','_') for str in columnas]
    return data
```

```python
elimina_guiones(data)
data.columns
```

```python
##IMPORTANTE: esta funcion solo se puede ejecutar una vez porque la compracion =='?' solo se puede hacer contra objetos
## y luego de ejecutada la funcin parasaria a ser numerico.
def eliminar_caracteres_a_numerico(data):
    import pandas as pd
    import numpy as np
    
    columnas = ['price', 'bore', 'stroke', 'horsepower', 'peak_rpm']
        
    for col in columnas:
        data.loc[data[col]=='?',col]=np.nan
        data.dropna(axis=0,inplace=True)
    
    for col in columnas:
        data[col] = pd.to_numeric(data[col])
    return data
    
```

```python
eliminar_caracteres_a_numerico(data)
```

```python
def crea_histograma_var_numericas(data, columns, bins = 10):
    for col in columns:
        fig = plt.figure(figsize=(6,6))
        ax = fig.gca()
        data[col].plot.hist(ax=ax, bins = bins)
        ax.set_title('Histograma para la columna: ' + col)
        ax.set_xlabel(col)
        ax.set_ylabel('cantidades para : ' + col)
        plt.show()
```

```python
columnas = ['curb_weight', 'engine_size', 'city_mpg', 'price'] 
crea_histograma_var_numericas(data,columnas)
```
