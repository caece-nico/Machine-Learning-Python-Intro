```python
import seaborn as sea
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'Automobile price data _Raw_.csv'
data = pd.read_csv(url)
data[1:5]
```

```python
##creamos la fucion para eliminar guiones y combertir a numerico lo ue es objeto

def limpia_datos(data):
    
    import numpy as np
    import pandas as pd
    
    columns = data.columns
    data.columns = [str.replace('-','_') for str in columns]
    
    ## buscamos los ? en los varables que queremos hacer numericas
    
    columns = ['price', 'bore', 'stroke', 'horsepower', 'peak_rpm']
    for col in columns:
        data.loc[data[col]=='?',col]=np.nan
        data[col].dropna(axis=0,inplace=True)
        
    for col in columns:
        data[col] = pd.to_numeric(data[col])
        
    return data    
```

```python
limpia_datos(data)
```

```python
data.dtypes
```

```python
## Creamos un grafico de estilo scatter plot para comprara varias variables numericas contra la variable precio.

def crea_scatter_plot(data, x_col, y_col = 'price'):
    for col in x_col:
        fig = plt.figure(figsize=(7,6))
        ax = fig.gca()
        data.plot.scatter(col, y_col, ax = ax)
        #ax.set_title('Relacion entre la variable: ' + y_col + ' vs ' + col)
        ax.set_xlabel(col)
        ax.set_ylabel(y_col)
        plt.show()
```

```python
num_cols = ['curb_weight', 'engine_size', 'horsepower', 'city_mpg']
crea_scatter_plot(data,num_cols)
```
