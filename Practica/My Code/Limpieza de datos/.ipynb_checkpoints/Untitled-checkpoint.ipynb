```python
import numpy as np
import pandas as pd
import seaborn as sea
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Automobile price data _Raw_.csv')
data.head(5)
```

```python
## quitamos los guiones

data.columns = [str.replace('-','_') for str in data.columns]
print(data.columns)
```

```python
### hacemos feature Eng cobre la columna. 

cylinder_categories = {'two':np.nan, 'three':'three_four', 'four':'three_four', 
                    'five':'five_six', 'six':'five_six',
                    'eight':'eight_twelve', 'twelve':'eight_twelve'}

data['num_of_cylinders'] = [cylinder_categories[x] for x in data['num_of_cylinders']]
print(data.head(20))
```

```python
def crea_boxplot(data, col, y_col = 'price'):
    for x_col in col:
        sea.set_style('whitegrid')
        sea.boxplot(x_col, y_col, data=data)
        plt.title('cc')
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.show()
```

```python
cols = ['price']

for c in cols:
    data.loc[data[c]=='?',c] = np.nan
    data.dropna(axis = 0, inplace = True)

for c in cols:
    data[c] = pd.to_numeric(data[c])

crea_boxplot(data,['num_of_cylinders'])
```

```python
## transformamos la variable body_cat
print(data['body_style'].value_counts())
body_cats = {'sedan':'sedan', 'hatchback':'hatchback', 'wagon':'wagon', 
             'hardtop':'hardtop_convert', 'convertible':'hardtop_convert'}

data['body_style'] = [body_cats[x] for x in data['body_style']]
print(data['body_style'].value_counts())
```

```python
crea_boxplot(data,['body_style'])
```

```python
### escalamos la vaiable Price a escala logaritmica.
def crea_histograma(var,lab, hist = False):
    sea.set_style('whitegrid')
    sea.distplot(var, hist = hist)
    plt.title(lab)
    plt.ylabel('densidad')
    plt.show()
    
crea_histograma(data['price'],'price', True)
```

```python
data['price_log'] = np.log(data['price'])
crea_histograma(data['price_log'],'price',True)
```

```python
###queda mucho mas normalizado.
```

```python
def crea_grafico_shapes(data, x_col,shape_col = 'fuel_type', y_col = 'price_log', alpha = 0.2 ):
    shape = ['+','*','x','o']
    categorias = data[shape_col].unique()
    for col in x_col:
        sea.set_style('whitegrid')
        for idx, cat in enumerate(categorias):
            temp = data[data[shape_col]==cat]
            sea.regplot(col, y_col, data = temp, marker = shape[idx], color = 'blue',label = cat,scatter_kws={'alpha':alpha}, fit_reg = False)
        plt.title('')
        plt.xlabel(col)
        plt.ylabel(y_col)
        plt.legend()
        plt.show()
```

```python
crea_grafico_shapes(data,['curb_weight', 'engine_size', 'horsepower', 'city_mpg'])
```
