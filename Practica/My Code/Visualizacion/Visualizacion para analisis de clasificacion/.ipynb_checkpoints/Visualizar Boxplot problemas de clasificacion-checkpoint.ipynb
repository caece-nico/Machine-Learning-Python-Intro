```python
"""
En este ejemplo creamos funciones para graficar distribuciones con bolxplot y barrras cuando lo que intentamos predecir es una
variable categorica(clasificacion) y NO es numerica(Regresion)

Boxpot: donde x_col es la variable categorica y y_col son los totales.
"""
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
url = 'German_Credit.csv'
columnas = ['customer_id',
                  'checking_account_status', 'loan_duration_mo', 'credit_history', 
                  'purpose', 'loan_amount', 'savings_account_balance', 
                  'time_employed_yrs', 'payment_pcnt_income','gender_status', 
                  'other_signators', 'time_in_residence', 'property', 'age_yrs',
                  'other_credit_outstanding', 'home_ownership', 'number_loans', 
                  'job_category', 'dependents', 'telephone', 'foreign_worker', 
                  'bad_credit']
data = pd.read_csv(url, header = None, names = columnas)
data.head(5)
```

```python
##creamos un grafico de boxplot para algunas variables categoricas para conocer mejor su distribucion.

def crea_box_plot(data, columns, x_col = 'bad_credit'):
    for col in columns:
        sea.set_style('whitegrid')
        #sea.boxplot(data[x_col], data[col]) // se puede hacer tambien como esta abajo. es mejor.
        sea.boxplot(x_col,col,data=data)
        plt.title('Muestra variable: ' + x_col +' vs: ' + col)
        plt.xlabel(x_col)
        plt.ylabel(col)
        plt.show()
        
```

```python
columnas = ['loan_duration_mo', 'loan_amount', 'payment_pcnt_income',
            'age_yrs', 'number_loans', 'dependents']
crea_box_plot(data, columnas)
```

```python
### creamos un grafico de violin usando la libreria seaborn.

def crea_violin_plot(data,columnas, x_col = 'bad_credit'):
    for col in columnas:
        sea.set_style('darkgrid')
        sea.violinplot(x_col,col,data = data)
        plt.title('Muestra columna : ' + col + ' vs: ' + x_col)
        plt.xlabel(x_col)
        plt.ylabel(col)
        plt.show()
```

```python
num_cols = ['loan_duration_mo', 'loan_amount', 'payment_pcnt_income',
            'age_yrs', 'number_loans', 'dependents']
crea_violin_plot(data,num_cols)
```
