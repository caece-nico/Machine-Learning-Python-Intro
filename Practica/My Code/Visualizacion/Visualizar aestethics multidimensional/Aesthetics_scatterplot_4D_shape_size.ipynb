```python
"""
En este ejemplo desarrollamos una funcion que grafica un scatter plot en 4 dimensiones.
la cuarta dimension es del tipo tamaño que puede ser tanto una variable categorica como numerica, pero lo mas normal es que sea 
numerica y debe tener un factor de normalizacion.

Este nuevo valor va dentro de sea.regplot(......., scatter_kws={'alpha':alpha, 's':normalizacion*col_size**2}
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
url = 'Automobile price data _Raw_.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
def limpia_datos(data):
    
    import numpy as np
    import pandas as pd
    
    columnas = data.columns
    data.columns = [str.replace('-','_') for str in columnas]
    
    columnas = ['price', 'bore', 'stroke', 
          'horsepower', 'peak_rpm']
    
    for col in columnas:
        data.loc[data[col]=='?',col]=np.nan
        data[col].dropna(axis=0,inplace=True)
        
    for col in columnas:
        data[col] = pd.to_numeric(data[col])
        
    return(data)
```

```python
limpia_datos(data)
```

```python
###creamos una funcion que grafica un scatter plor en 4 dimensiones. con variable para formas y otra para tamaño (size)

def crea_scatter_4d(dataset, x_col, y_col = 'price',shape_col = 'fuel_type', size_col = 'curb_weight', normalizacion = 0.0000075, alpha = 0.3):
    shape = ['+','o','*','x']
    unique_categorias = dataset[shape_col].unique()
    for col in x_col:
        sea.set_style('whitegrid')
        for idx, cat in enumerate(unique_categorias):
            temp = dataset[dataset[shape_col]==cat]
            sea.regplot(col, y_col, temp, marker = shape[idx], label = cat, 
                       scatter_kws={'alpha':alpha, 's':normalizacion*temp[size_col]**2}, fit_reg = False,color = 'blue')
        plt.legend()
        plt.title('scatter plot of : ' + col + ' vs: ' + y_col)
        plt.xlabel(col)
        plt.ylabel(y_col)
        plt.show()
```

```python
columnas =  ['engine_size', 'horsepower', 'city_mpg']
crea_scatter_4d(data,columnas)
```
