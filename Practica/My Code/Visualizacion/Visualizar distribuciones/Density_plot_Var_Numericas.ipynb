```python
import seaborn as sea
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
url = 'Automobile price data _Raw_.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
data.dtypes
```

```python
def elimina_guiones_pasa_a_numerico(data):
    import pandas as pd
    import numpy as np
    
    col = data.columns
    data.columns = [str.replace('-','_') for str in col]
        
    
    columns = ['price', 'bore', 'stroke', 'horsepower', 'peak_rpm']
    
    for col in columns:
        data.loc[data[col]=='?',col] = np.nan
        data[col].dropna(axis=0,inplace=True)
        
    for col in columns:
        data[col] = pd.to_numeric(data[col])
        
    return data   
        
```

```python
elimina_guiones_pasa_a_numerico(data)
```

```python
def crea_density_plot_var_numerica(data,columns,bins = 10, hist = False):
    for col in columns:
        sea.set_style('whitegrid')
        sea.distplot( data[col],bins = bins, hist = hist, rug = True )
        plt.title('Distribucion para la variable : ' + col)
        plt.xlabel(col)
        plt.ylabel('Valores de : '+col)
        plt.show()
        
```

```python
columnas = ['curb_weight', 'engine_size','city_mpg', 'price'] 
crea_density_plot_var_numerica(data,columnas)
```

```python
## Al density plot anterior le podemos agregar un histograma interno para ver las informacion
crea_density_plot_var_numerica(data,columnas, hist = True)
```
