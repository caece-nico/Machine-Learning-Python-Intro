```python
"""
En este notebook creamos una funcion para graficar violin usando seaborn.
la idea es poder graficar una variable numerica contra una variable categorica abriendola en sus componentes.
y muestra la densidad de la variable categorica.
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
url = 'Automobile price data _Raw_.csv'
data = pd.read_csv(url)
data.head(5)
```

```python
## cramos una funcion que elimina los guiones del titulo.
def elimina_guiones(data):
    columnas = data.columns
    data.columns = [str.replace('-','_') for str in columnas]
```

```python
elimina_guiones(data)
data.columns
```

```python
## creamos una funcion que eliminar faltantes y convierte a numerico

def convierte_a_numerico(data):
    import numpy as np
    import pandas as pd
    
    columnas = ['price', 'bore', 'stroke', 
          'horsepower', 'peak_rpm']
    
    for col in columnas:
        data.loc[data[col]=='?', col] = np.nan
        data[col].dropna(axis=0,inplace=True)
    
    for col in columnas:
        data[col]= pd.to_numeric(data[col])
        
    return data
```

```python
convierte_a_numerico(data)
data.dtypes
```

```python
##creamos una funcione que grafica un boxplot de variables categoricas contra una variable numerico = 'Precio'

def crea_violinplot(data, x_col,y_col = 'price'):
    for col in x_col:
        sea.set_style('whitegrid')
        sea.violinplot(col,y_col,data = data)
        plt.title('Categorica: ' + col + ' Numerica: ' + y_col)
        plt.xlabel(col)
        plt.ylabel(y_col)
        plt.show()
```

```python
columnas = ['fuel_type', 'aspiration', 'num_of_doors', 'body_style', 
            'drive_wheels', 'engine_location', 'engine_type', 'num_of_cylinders']
crea_violinplot(data,columnas)
```

```python
data[data['num_of_doors']=='?'] = np.nan
##Practica:
## como hay dos autos que no tienen cantidad de puertas las eliminamos.

data['num_of_doors'].dropna(axis=0, inplace=True)
```

```python
crea_violinplot(data,['num_of_doors'])
```
